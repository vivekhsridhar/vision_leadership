```python
import numpy as np
import pandas as pd
from keras.layers import Input, Conv1D, AveragePooling1D, UpSampling1D, Add
from keras.models import Model
from keras.optimizers import Adam
```

```python
foldername = '0066'
```

```python
df = pd.read_hdf('/home/user/Documents/Vivek/dnn/vision_leadership/golden_shiners/10_fish/' + foldername + '/' + foldername + '.h5')
df.head()
```

```python
padding = 'causal'

def conv_pool(conv_in, nfilters):
    conv = Conv1D(filters=nfilters, kernel_size=3, padding=padding, activation='selu', use_bias=False)(conv_in)
    pool = AveragePooling1D(pool_size=2)(conv)
    return conv, pool

def upsample_add_deconv(upsamp_in, res, nfilters):
    upsample = UpSampling1D(size=2)(upsamp_in)
    add = Add()([res,upsample])
    deconv = Conv1D(filters=nfilters, kernel_size=3, padding=padding, activation='selu', use_bias=False)(add)
    return upsample, add, deconv
```

```python
input_layer = Input(batch_shape=(None, None, 6))

# --------------------------------- bottom-up part of the hourglass --------------------------------- #
conv1, pool1 = conv_pool(input_layer, 32)
conv2, pool2 = conv_pool(pool1, 64)
conv3, pool3 = conv_pool(pool2, 128)
conv4, pool4 = conv_pool(pool3, 256)
conv5 = Conv1D(filters=512, kernel_size=3, padding=padding, activation='selu', use_bias=False)(pool4)
# --------------------------------- xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx --------------------------------- #

# ------------- modify conv layer output to match shape to corresponding upsample layer ------------- #
res4 = Conv1D(filters=512, kernel_size=3, padding=padding, activation='selu', use_bias=False)(conv4)
res3 = Conv1D(filters=256, kernel_size=3, padding=padding, activation='selu', use_bias=False)(conv3)
res2 = Conv1D(filters=128, kernel_size=3, padding=padding, activation='selu', use_bias=False)(conv2)
res1 = Conv1D(filters=64, kernel_size=3, padding=padding, activation='selu', use_bias=False)(conv1)
# --------------------------------- xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx --------------------------------- #

# --------------------------------- top-down part of the hourglass ---------------------------------- #
upsample4, add4, deconv4 = upsample_add_deconv(conv5, res4, 256)
upsample3, add3, deconv3 = upsample_add_deconv(deconv4, res3, 128)
upsample2, add2, deconv2 = upsample_add_deconv(deconv3, res2, 64)
upsample1, add1, deconv1 = upsample_add_deconv(deconv2, res1, 32)
# --------------------------------- xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx --------------------------------- #

output_layer = Conv1D(filters=2, kernel_size=3, padding=padding, activation='linear')(deconv1)

model = Model(input_layer, output_layer)
model.compile(Adam(lr=0.001, amsgrad=True), loss='mse')
```

```python
half_batch = 4096
f_id = 7
n_id = 8
def generate_batch():
    #f_id = np.random.choice(np.arange(min(df['f_id']),max(df['f_id'])+1), 1)[0]
    #n_id = np.random.randint(min(df['n_id']),max(df['n_id'])+1)
    tmp = df[(df['f_id'] == f_id) & (df['n_id'] == n_id)]
    
    idx = np.random.choice(half_batch,tmp.shape[0]-half_batch), 1)
    x = tmp[idx-half_batch:idx+half_batch][['ang_area_left_eye', 'ang_area_right_eye', 'cos_pos_left_eye', 'sin_pos_left_eye', 'cos_pos_right_eye', 'sin_pos_right_eye']].values[None,...]
    y = tmp[idx-half_batch:idx+half_batch][['cc', 'tau']].values[None,...]
    
    return x,y

def generator():
    while(True):
        x,y = generate_batch()
        
        yield (x,y)
```

```python
model.fit_generator(generator(), 1000, 10, workers=8, use_multiprocessing=True)
```

```python
x,y = generate_batch()

y_preds = model.predict(x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.scatter(y_preds[0,...,0], y[0,...,0], alpha=0.5)
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
plt.plot(y[0,...,0], 'k-', alpha=0.8)
plt.plot(y_preds[0,...,0], 'r-', alpha=0.5)
plt.show()
```
